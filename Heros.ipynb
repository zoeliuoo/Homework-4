```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data
```

```python
count_players = purchase_data['SN'].nunique()
count_players
df1 = pd.DataFrame([[count_players]],columns=['Total Players'])
df1
```

```python
count_id = purchase_data['Item ID'].nunique()
ave_price = purchase_data['Price'].mean()
purchase_count = purchase_data['SN'].count()
Total_Revenue = purchase_data['Price'].sum()

df2 = pd.DataFrame({'Number of Unique Items':[count_id],'Average Price':[ave_price],'Number of Purchases':[purchase_count],'Total Revenue':[Total_Revenue]})
df2
```

```python
unique_player= purchase_data.drop_duplicates(subset='SN', keep='first', inplace=False)
value_gender = unique_player['Gender'].value_counts()
percentage =value_gender/count_players

df3 = pd.DataFrame(list(zip(value_gender, percentage)), 
               columns =['Value', 'Percentage'],index=['Male','Female','Others/ Non-Disclosed']) 
df3 
```

```python
groups = purchase_data.groupby(['Gender'])
purchase_count = groups['SN'].count()
Ave_price = groups['Price'].mean()
Total_price = groups["Price"].sum()
a = Total_price/value_gender

df4 = pd.DataFrame(list(zip(purchase_count,Ave_price,Total_price,a)), 
               columns =['Purchase Count', 'Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person'],index=['Male','Female','Others/ Non-Disclosed']) 
df4 
```

```python
age = unique_player.iloc[:,2]
bins=[0,9,14,19,24,29,34,39,100]
cats =pd.cut(age,bins)

age_group=pd.value_counts(cats)
percentage_age = age_group/576

df5 = pd.DataFrame(list(zip(age_group, percentage_age)), 
                   columns =['Total Value', 'Percentage'],index=['20-24','15-19','25-29','30-34','35-39','10-14','<10','40+']) 
new_index= ['<10', '10-14', '15-19', '20-24','25-29','30-34','35-39','40+']
df5 = df5.reindex(new_index)
df5
```

```python
l1 = purchase_data.loc[purchase_data["Age"] < 10]
purchase_count_1 = l1['SN'].count()
Ave_price_1 = l1['Price'].mean()
Total_price_1 = l1["Price"].sum()
a_1 = Total_price_1/l1['SN'].nunique()
row1 = [purchase_count_1,Ave_price_1,Total_price_1,a_1]

l2 = purchase_data.loc[(purchase_data["Age"] >= 10) & (purchase_data["Age"]<15)]
purchase_count_2 = l2['SN'].count()
Ave_price_2 = l2['Price'].mean()
Total_price_2 = l2["Price"].sum()
a_2 = Total_price_2/l2['SN'].nunique()
row2 = [purchase_count_2,Ave_price_2,Total_price_2,a_2]

l3 = purchase_data.loc[(purchase_data["Age"] >= 15) & (purchase_data["Age"]<20)]
purchase_count_3 = l3['SN'].count()
Ave_price_3 = l3['Price'].mean()
Total_price_3 = l3["Price"].sum()
a_3 = Total_price_3/l3['SN'].nunique()
row3 = [purchase_count_3,Ave_price_3,Total_price_3,a_3]

l4 = purchase_data.loc[(purchase_data["Age"] >= 20) & (purchase_data["Age"]<25)]
purchase_count_4 = l4['SN'].count()
Ave_price_4 = l4['Price'].mean()
Total_price_4 = l4["Price"].sum()
a_4 = Total_price_4/l4['SN'].nunique()
row4 = [purchase_count_4,Ave_price_4,Total_price_4,a_4]

l5 = purchase_data.loc[(purchase_data["Age"] >= 25) & (purchase_data["Age"]<30)]
purchase_count_5 = l5['SN'].count()
Ave_price_5 = l5['Price'].mean()
Total_price_5 = l5["Price"].sum()
a_5 = Total_price_5/l5['SN'].nunique()
row5 = [purchase_count_5,Ave_price_5,Total_price_5,a_5]

l6 = purchase_data.loc[(purchase_data["Age"] >= 30) & (purchase_data["Age"]<35)]
purchase_count_6 = l6['SN'].count()
Ave_price_6 = l6['Price'].mean()
Total_price_6 = l6["Price"].sum()
a_6 = Total_price_6/l6['SN'].nunique()
row6 = [purchase_count_6,Ave_price_6,Total_price_6,a_6]

l7 = purchase_data.loc[(purchase_data["Age"] >= 35) & (purchase_data["Age"]<40)]
purchase_count_7 = l7['SN'].count()
Ave_price_7 = l7['Price'].mean()
Total_price_7 = l7["Price"].sum()
a_7 = Total_price_7/l7['SN'].nunique()
row7 = [purchase_count_7,Ave_price_7,Total_price_7,a_7]

l8 = purchase_data.loc[(purchase_data["Age"] >= 40)]
purchase_count_8 = l8['SN'].count()
Ave_price_8 = l8['Price'].mean()
Total_price_8 = l8["Price"].sum()
a_8 = Total_price_8/l8['SN'].nunique()
row8 = [purchase_count_8,Ave_price_8,Total_price_8,a_8]

df6 = pd.DataFrame([row1,row2,row3,row4,row5,row6,row7,row8]
                   ,index=['<10','10-15','15-20','20-25','25-30','30-35','35-40','40+']
                   ,columns =['Purchase Count','Average Purchase Price','Total Purchase Value','Avg Total Purchase per Person'])
df6
```
